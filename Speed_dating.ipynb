```python
# I prefer to add a unique id for each candidate , this will be useful for future use,
# this unique id is comprised of the wave number, the candidate id number in this wave, and male or female

# assign unique id based on values of columns
def assign_unique_id(row):

  # extracting male or female
  if row['gender'] == 0:
    gender = 'F'
  else:
    gender = 'M'

  unique_id =  gender + str(row['wave']) + '_' + str(row['id'])
  return unique_id

# create the unique_id column
df['unique_id'] = df.apply(assign_unique_id, axis=1)  # don't forget (axis = 1) to differentiate between column or rows direction not rows
```

### C.2) ATTRIBUTES OBSERVER

Establishing a petite dataframe for the results and *charts* especially designed to be use dby plotly

```python
# two dataframes one that looks like a table for reporting and troubleshooting
df_attributes = pd.DataFrame(columns = ['gender','status','Atractive','Sincere','Intelligent','Fun','Ambitious','Shared_interest','flag'])

# the second is designed to be used by plotly
df_graph = pd.DataFrame(columns = ['gender','attribute','value','status'])
```

```python
# This function will calcualte all attributes and produce a table report
# And another graph dataset especially to use very simply plotly features


def calculate_attributes(scope_to_cover,scope_name):
  print('Report started')
  for scope, name in zip(scope_to_cover,scope_name):

    # 1) preparing the data
    needed_columns = scope.copy()     # extracting the needed attributes
    needed_columns = needed_columns + ['gender','unique_id']  # adding gender and unique_id
    df_lab = df.loc[:, needed_columns].copy()  # masking all the mentioned
    # try the above with df_match and df_no_match they all produce the same results

    print(f'Calculating for attribute "{name}":')
    print(f' dataframe shape is {df_lab.shape}')
    print()

    print(' dropping rows with empty fields...' )
    df_lab = df_lab.dropna()

    print(' Data shape after dropping the empty fields:' ,df_lab.shape)
    print()

    ## === Attention === ###
    ## There are duplicate lines becasue these attributes are repeated for each interview
    print(' Checking duplicates and dropping them..')
    df_lab = df_lab.drop_duplicates()
    print('After dropping', df_lab.shape)
    print()

    # 2) Calculate
    grouped_res = df_lab.groupby('gender')[scope].mean()
    display(grouped_res)

    # 3) Inserting into report table format
    # 3.1) define a new item to insert in attribute_dataframe for reporting
    # why? because we want it like offical report
    for i in range(2):
      new_item = {'gender': i,
                  'status': name,
                  'Atractive': grouped_res.iloc[i,0],
                  'Sincere':grouped_res.iloc[i,1],
                  'Intelligent':grouped_res.iloc[i,2],
                  'Fun':grouped_res.iloc[i,3],
                  'Ambitious':grouped_res.iloc[i,4],
                  'Shared_interest':grouped_res.iloc[i,5],
                  'flag': 0 }

      # 3.2 )define a pointer at the end of the df_attribute to insert data
      index_df_attributes = df_attributes.shape[0]

      # 3.3) Insert data
      df_attributes.loc[index_df_attributes] = new_item

      # 4 ) graph table to benefit from all plotly features
      attribute_names = ['Attractive','Sincere','Intelligent','Fun','Ambitious','Shared_interest']
      for attribute_index in range(5):
        graph_entity = {'gender':i,
                      'attribute': attribute_names[attribute_index],
                        'value': grouped_res.iloc[i,attribute_index],
                        'status': name
                        }

        # define graph dataset indicator to insert
        index_df_graph = df_graph.shape[0]

        # Insert
        df_graph.loc[index_df_graph] = graph_entity



```

## C.5) Rating analysis
There are two ratings
First is the personal rating of the particpant to himself

Second the the ratings of the partners who met with him.

How compatible are these results together, are the particpants able to evaluate themselves right? are the partners evaluation accurate?

bearing in mind that **I CHOSE ON PURPOSE TO ANALYZE THE CASES WHERE A MATCH OCCURRED.**

```python
## APOLOGIES THIS PART ONLY OF CODE IS NOT DRY, BECAUSE LIMITED IN TIME
r1 = ['attr3_1','sinc3_1','intel3_1', 'fun3_1','amb3_1']
r2 = ['attr3_2','sinc3_2','intel3_2', 'fun3_2','amb3_2']
r3 = ['attr_o','sinc_o','intel_o', 'fun_o','amb_o']

df_ratings = pd.DataFrame(
    columns=['unique_id','gender', 'attribute', 'value','status']
)

for col in r1:

  needed_columns = ['unique_id','gender',col,'partner']
  bring = df_match.loc[:,needed_columns]

  # Get the existing column names
  column_names = bring.columns.tolist()

  # Rename the third column
  column_names[2] = 'value'

  bring.columns = column_names

  bring['status'] = 'evaluate_yourself'
  bring['attribute'] = col

  df_ratings = pd.concat([df_ratings, bring])
#############################
for col in r2:

  needed_columns = ['unique_id','gender',col,'partner']
  bring = df_match.loc[:,needed_columns]

  # Get the existing column names
  column_names = bring.columns.tolist()

  # Rename the third column
  column_names[2] = 'value'

  bring.columns = column_names

  bring['status'] = 'evaluate_yourself_after3weeks'
  bring['attribute'] = col

  df_ratings = pd.concat([df_ratings, bring])
###############################

for col in r3:

  needed_columns = ['unique_id','gender',col,'partner']
  bring = df_match.loc[:,needed_columns]

  # Get the existing column names
  column_names = bring.columns.tolist()

  # Rename the third column
  column_names[2] = 'value'

  bring.columns = column_names

  bring['status'] = 'rate_by_partner'
  bring['attribute'] = col

  df_ratings = pd.concat([df_ratings, bring])

def condition(x):
    if 'attr' in x:
        return "Attractive"
    elif 'sinc' in x :
        return "Sincere"
    elif 'intel' in x:
        return 'Intelligent'
    elif 'fun' in x:
        return 'Funny'
    elif 'amb' in x:
        return 'Ambitious'
    else :
        return 'NA'

# Applying the conditions
df_ratings['attribute'] = df_ratings['attribute'].apply(condition)
```

```python
mask = df_ratings['unique_id'].str.contains("F1.1")
#mask = (df_ratings['unique_id'] == 'F1.1') | (df_ratings['unique_id'] == 'F1.2')

#mask = df_ratings['unique_id'] == 'F1_1.0'
df_chart = df_ratings.loc[mask,:]
fig= px.bar(df_chart, x='status', color='attribute', y='value', title="Perception Evolution with time", barmode='group',
            facet_row='unique_id',
            facet_col='partner')
fig.show()
```
